```python
comanpy = soup.find_all('div',class_="JobListCardstyles__JobCardWrapper-ffng7u-0 liPDyT")
```

```python
job_titel =[]
company_name = []
Loction =[]
salery = []
Other_role = []


for i in comanpy:
    job_titel.append(i.find('div',class_="JobListCardstyles__JobTitle-ffng7u-7 cuaBGE").text.strip())
    company_name.append(i.find('div',class_="JobListCardstyles__JobCompany-ffng7u-8 gguURM").text.strip())
    Loction.append(i.find_all('div',class_="JobListCardstyles__DisplayFlexCenter-ffng7u-10 fNeylV")[0].text.strip())
    salery.append(i.find_all('div',class_="JobListCardstyles__DisplayFlexCenter-ffng7u-10 fNeylV")[1].text.strip())
    other_role=(i.find('div',class_="JobListCardstyles__TagContainerWrapper-ffng7u-12 bvKvxY").text.strip())
    cleaned_text = other_role.replace('\xa0', ' ')
    Other_role.append(cleaned_text)
    
    
df = pd.DataFrame({
    "job_titel":job_titel,
    "company_name":company_name,
    "Loction":Loction,
    "salery" : salery,
    "other_role":Other_role
})
df.shape
```

```python
final = pd.DataFrame()
for j in range (1,371):
    url = "https://apna.co/jobs/freshers-jobs?page={}".format(j)
    webpage = requests.get(url).text
    soup= BeautifulSoup(webpage,'lxml')
    comanpy = soup.find_all('div',class_="JobListCardstyles__JobCardWrapper-ffng7u-0 liPDyT")
    job_titel =[]
    company_name = []
    Loction =[]
    salery = []
    Other_role = []
    for i in comanpy:
        try:
            job_titel.append(i.find('div',class_="JobListCardstyles__JobTitle-ffng7u-7 cuaBGE").text.strip())
        except:
            job_titel.append(np.nan)
        try:
            company_name.append(i.find('div',class_="JobListCardstyles__JobCompany-ffng7u-8 gguURM").text.strip())
        except:
            company_name.append(np.nan)
        try:
            Loction.append(i.find_all('div',class_="JobListCardstyles__DisplayFlexCenter-ffng7u-10 fNeylV")[0].text.strip())
        except:
            Loction.append(np.nan)
        try:
            salery.append(i.find_all('div',class_="JobListCardstyles__DisplayFlexCenter-ffng7u-10 fNeylV")[1].text.strip())
        except:
            salery.append(np.nan)
        try:
            other_role=(i.find('div',class_="JobListCardstyles__TagContainerWrapper-ffng7u-12 bvKvxY").text.strip())

            cleaned_text = other_role.replace('\xa0', ' ')
            Other_role.append(cleaned_text)
        except:
            Other_role.append(np.nan)
            
    df = pd.DataFrame({
    "job_titel":job_titel,
    "company_name":company_name,
    "Loction":Loction,
    "salery" : salery,
    "other_role":Other_role})
    final=final.append(df,ignore_index=True)
```
